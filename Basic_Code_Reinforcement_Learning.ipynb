## Target: Maximizing Rewards

```python
import random
from typing import List

class Environment:
    def __init__(self):
        self.steps_left = 20

    def get_observation(self) -> List[float]:
        return [0.0, 0.0, 0.0]

    def get_actions(self) -> List[int]:
        return [0, 1]

    def is_done(self) -> bool:
        return self.steps_left == 0

    def action(self, action: int) -> float:
        if self.is_done():
            raise Exception("The game is over!")
        self.steps_left -= 1
        return random.random()
```

```python
random.choice([0,1])
```

```python
class Agent:
    def __init__(self):
        self.total_reward = 0.0

    def step(self, env: Environment):
        current_obs = env.get_observation()
        print("Observation {}".format(current_obs))
        actions = env.get_actions()
        print(actions)
        reward = env.action(random.choice(actions))
        self.total_reward += reward
        print("Total Reward {}".format(self.total_reward))
```

```python
if __name__ == "__main__":
    env = Environment()
    agent = Agent()
    i=0

    while not env.is_done():
        i=i+1
        print(f"Steps: {i}")
        agent.step(env)

    print("Your cumulative reward is: %.4f" % agent.total_reward)
```

## Within the steps (here we fixed 20 steps), the agent get a total reward in each step corresponding to the environment (through random iteration)!
